```python
#tratamiento de datos nulos 
df_sinnulos=salario.dropna()
df_sinulos_columnas=salario.dropna(axis=1)
df_sinnulos
```

```python
df_sinnulos['salary']
```

```python
## Este cÃ³digo toma una columna `'salary'` que contiene valores de salario en formato de cadena, 
## elimina cualquier caracter no numÃ©rico, extrae los dÃ­gitos y convierte la columna resultante en valores numÃ©ricos de tipo float. 
## El resultado se almacena en una nueva columna llamada `'salary_numeric'`. 
## Esto es Ãºtil cuando la columna `'salary'` inicial contiene caracteres no numÃ©ricos y se desea trabajar con los salarios como valores numÃ©ricos
df_sinnulos['salary_numeric']  = df_sinnulos['salary'].replace('[^\d.]', '', regex=True).str.extract('(\d+)').astype(float)

#que tipo de funciones aplicar para parsear la columna
```

```python
df_sinnulos['salary_numeric'].describe
```

```python
df_sinnulos.drop('salary',axis=1,inplace=True)
```

```python
df_sinnulos['salary_numeric']=df_sinnulos['salary_numeric'].round(0).astype(int)
```
